<h3> Exploratory data analysis (EDA)

- <h4>what is the overall sales trend?

```python
print(df['order_date'].min())
print(df['order_date'].max())
```

```python
#add month year
df['month_year'] = df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
df['month_year']
```

```python
df.info()
```

```python
df['order_date'] = df['order_date'].astype('int64')
df['ship_date'] = df['ship_date'].astype('int64')
df_month_group = df.groupby('month_year').sum()['sales'].reset_index()
```

```python
df_month_group
```

```python
#setting the figure size
plt.figure(figsize=(16, 5))
plt.plot(df_month_group['month_year'], df_month_group['sales'], color = '#b80045')
plt.xticks(rotation = 'vertical', size = 10)
plt.show()
```

<h4> 2. Find top 10 products by sales

```python
#groupping products by sales
df_prod_sales = pd.DataFrame(df.groupby('product_name').sum()['sales'])

#sort by sales
df_prod_sales.sort_values(by = ['sales'], inplace = True, ascending = False)

#top10 products by sales
df_prod_sales[:10]
```

<h4>3. What is the most preferred ship mode?

```python
#setting the figue size
plt.figure(figsize=(10, 8)) 

#  countplot
sns.countplot(x = 'ship_mode', data = df)

plt.show()
```

<h4> 4. WHICH ARE THE MOST PROFITABLE CATEGORY AND SUB-CATEGORY?

```python
#groupping
cat_subcat = pd.DataFrame(df.groupby(['category', 'sub_category']).sum()['profit'])
cat_subcat.sort_values(['category', 'profit'], ascending= False)
```
